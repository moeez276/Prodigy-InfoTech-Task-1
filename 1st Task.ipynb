```python
# Drop unuseful columns
df.drop(columns=['Country Code','Indicator Name'],inplace=True)
```

```python
# filter data for total population
total_population_df=df[df['Indicator Code']=='SP.POP.TOTL']
total_population_df
```

```python
# sort data based on total population for 2022
total_population_sort = total_population_df.sort_values(by='2022',ascending=False)
total_population_sort
```

```python
# get top 10 countries with the highest population for 2022
top_ten_countries = total_population_sort.head(10)
print("Top Ten Counteies Of Total Population:-")
print(top_ten_countries)
```

```python
# Data visualization
def add_value_labels(ax,spacing=10):
    for rect in ax.patches:
        x=rect.get_x()+rect.get_width()/2
        y=rect.get_height()-3
        
        va='bottom' if y>=0 else 'top'
        label= '()'.format(y) 
        y_shift = spacing * (1 if y >= 0 else -1)
        ax.annotate(label,(x,y),xytext=(0,y_shift),textcoords="offset points",ha='center',va=va)
```

```python
# Assuming top_ten_countries is your DataFrame
plt.figure(figsize=(10, 6))

# For the 2022 barplot
plot = sns.barplot(x='2022', y='Country Name', data=top_ten_countries, hue='Country Name', palette='coolwarm', legend=False)
plt.title('Top Ten Countries in 2022')
plt.show()

# For the 2000 barplot
plt.figure(figsize=(10, 6))
plot = sns.barplot(x='2000', y='Country Name', data=top_ten_countries, hue='Country Name', palette='coolwarm', legend=False)
plt.title('Top Ten Countries in 2000')
plt.show()
```

```python
# sort data based on total population for 2022
total_population_sort_bt = total_population_df.sort_values(by='2022',ascending=True)
total_population_sort_bt
```

```python
top_ten_bt_countries = total_population_sort_bt.head(10)
print("Top Ten Bottom Counteies Of Total Population:-")
print(top_ten_bt_countries)
```

```python
# Create a figure with two subplots, stacked vertically (1 column, 2 rows)
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 12))

# Plot for 2022
sns.barplot(x='2022', y="Country Name", data=top_ten_bt_countries, hue='Country Name', palette="coolwarm", ax=axes[0], legend=False)
axes[0].set_title("Top Ten Bottom Countries of Total Population (2022)")
axes[0].set_xlabel("Total Population")
axes[0].set_ylabel("Country")

# Plot for 2000
sns.barplot(x='2000', y="Country Name", data=top_ten_bt_countries, hue='Country Name', palette="coolwarm", ax=axes[1], legend=False)
axes[1].set_title("Top Ten Bottom Countries of Total Population (2000)")
axes[1].set_xlabel("Total Population")
axes[1].set_ylabel("Country")

# Adjust layout
plt.tight_layout()
plt.show()
```
